```python
import json 
import os 
```

```python
path = 'E:/22.9.21/GM/openpose/custom_dataset/annotations'
valid_path = os.path.join(path, 'valid_baby_keypoints.json')
```

```python
with open(valid_path, 'r') as f:
    json_data = json.load(f)
```

```python
coco = [] 
coco_data = {}
for ann in json_data['annotations']:
    # print(f"------------------->{ann['image_id']}")
    # print(f"'------------------->{ann['bbox']}")
    # print(f"'------------------->{ann['category_id']}")
    coco_data['bbox'] = ann['bbox']
    coco_data['category_id'] = ann['category_id']
    coco_data['image_id'] = ann['image_id']
    coco_data['score'] = 1
    
    coco.append(coco_data)

coco
    
```

```python
with open(os.path.join(path, 'valid_bbox.json'), 'w') as f:
        json.dump(coco, f)
```
